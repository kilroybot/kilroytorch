# Multiclass Example

##### Imports

```python
import warnings
from tqdm import TqdmWarning
warnings.filterwarnings("ignore", category=TqdmWarning)
```

```python
import numpy as np
import torch
from torch import nn, optim
from tqdm import tqdm
from kilroyplot.plot import plt
from kilroyshare.codec import Codec
from kilroytorch.modules.offline.multiclass import SimpleMulticlassOfflineModule as OfflineModule
from kilroytorch.modules.online.multiclass import SimpleMulticlassOnlineModule as OnlineModule
from kilroytorch.models.distribution.multiclass import LinearMulticlassDistributionModel as Model
```

##### Helpers

```python
def class_bins(n_classes):
    return np.arange(-0.5, n_classes+0.5)
    
def chunks(data, size):
    for i in range(0, len(data), size):
        yield data[i:i + size]

def generate(model, codec, n):
    return [codec.encode(sample) for sample in model.sample(n)[0]]

def proportional_density(support, f):
    values = np.array([f(x) for x in support])
    weights = values - values.min()
    return weights / weights.sum() if weights.sum() > 0 else np.ones(len(weights)) / len(weights)
```

## Task

Let's assume the values of the score function for all classes are like shown below.

```python
n_classes = 300
```

```python
score = lambda value: np.sin(6 * np.pi * value / (n_classes-1)) + np.random.randn() * 0.1  # sine plus some noise
```

```python
plt.plot(range(n_classes), [score(c) for c in range(n_classes)])
plt.title("Exemplary score function")
plt.xlabel("class")
plt.ylabel("score")
plt.tight_layout()
```

## Model

```python
class DummyCodec(Codec):
    def encode(self, value):
        return value.item()

    def decode(self, value):
        return torch.tensor(value)
```

```python
model = Model(n_classes)
codec = DummyCodec()
offline_module = OfflineModule(model, codec, optim.Adam(model.parameters(), lr=0.05))
online_module = OnlineModule(model, codec, optim.Adam(model.parameters(), lr=0.05))
```

## Offline

Let's assume real samples are distributed proportional to their relative scores.

```python
probs = proportional_density(range(n_classes), score)
n_samples = 1000
samples = np.random.choice(n_classes, n_samples, p=probs)
```

```python
plt.hist(samples, bins=class_bins(n_classes))
plt.title("Histogram of samples")
plt.xlabel("class")
plt.ylabel("frequency")
plt.tight_layout()
```

```python
n_epochs = 100
batch_size = 512
losses = []

for _ in tqdm(range(n_epochs)):
    epoch_losses = []
    for batch in chunks(samples, batch_size):
        metrics = offline_module.fit(batch)
        epoch_losses.append(metrics["loss"])
        offline_module = offline_module.step()
    losses.append(np.mean(epoch_losses))
```

```python
plt.plot(losses)
plt.title("Mean loss each epoch")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.tight_layout()
```

```python
plt.hist(generate(model, codec, 1000), bins=class_bins(n_classes))
plt.title("Histogram of generated samples")
plt.xlabel("class")
plt.ylabel("frequency")
plt.tight_layout()
```

## Online

```python
n_iters = 100
iter_size = 512
losses = []
mean_scores = []

for _ in tqdm(range(n_iters)):
    scores = {
        key: score(value) for key, value in online_module.sample(iter_size)
    }
    mean_scores.append(np.mean(list(scores.values())))
    metrics = online_module.fit(scores)
    losses.append(metrics["loss"])
    online_module = online_module.step()
```

```python
plt.plot(losses)
plt.title("Loss each iteration")
plt.xlabel("iteration")
plt.ylabel("loss")
plt.tight_layout()
```

```python
plt.plot(mean_scores)
plt.title("Mean scores of generated samples")
plt.xlabel("iteration")
plt.ylabel("mean score")
plt.tight_layout()
```

```python
plt.hist(generate(model, codec, 1000), bins=class_bins(n_classes))
plt.title("Histogram of generated samples")
plt.xlabel("class")
plt.ylabel("frequency")
plt.tight_layout()
```
